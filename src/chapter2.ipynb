```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
heights = []

with open("./data/heights.txt", "r") as f:
    for line in f:
        heights.append(float(line.strip()))
```

```python
heights[:5]
```

```python
def normal(x: np.ndarray, mu: float, sigma: float) -> np.ndarray:
    return np.exp(-((x - mu) ** 2) / (2 * sigma**2)) / (np.sqrt(2 * np.pi) * sigma)


def standard_normal(x: np.ndarray) -> np.ndarray:
    return normal(x, 0, 1)
```

```python
mu = float(np.mean(heights))
sigma = float(np.std(heights))

print(f"Mean: {mu}, Standard Deviation: {sigma}")
```

```python
x = np.linspace(np.min(heights), np.max(heights), 1000)
y = normal(x, mu, sigma)

plt.hist(
    heights, bins="auto", density=True, alpha=0.5, color="blue", label="Height Data"
)
plt.plot(x, y, color="red")
plt.xlabel("Height (cm)")
plt.ylabel("Probability Density")
plt.title("Height Distribution")
plt.show()
```

```python
samples = [np.random.normal(mu, sigma) for _ in range(len(heights))]

plt.hist(
    samples, bins="auto", density=True, alpha=0.5, color="orange", label="Sample Data"
)
plt.hist(
    heights, bins="auto", density=True, alpha=0.5, color="blue", label="Height Data"
)
plt.legend()
plt.show()
```

```python
print("<160", scipy.stats.norm.cdf(160, loc=mu, scale=sigma))
print(">180", 1.0 - scipy.stats.norm.cdf(180, loc=mu, scale=sigma))
```
