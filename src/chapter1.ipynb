```python
import numpy as np
import pandas as pd
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def normal(x: np.ndarray, mu: float, sigma: float) -> np.ndarray:
    return np.exp(-((x - mu) ** 2) / (2 * sigma**2)) / (np.sqrt(2 * np.pi) * sigma)


def standard_normal(x: np.ndarray) -> np.ndarray:
    return normal(x, 0, 1)
```

```python
x = np.linspace(-5, 5, 100)
y = standard_normal(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("ùí©(Œº=0, œÉ=1)")
plt.title("Standard Normal Distribution")
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y1 = standard_normal(x)
y2 = normal(x, -2, 1)
y3 = normal(x, 2, 1)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel("x")
plt.ylabel("ùí©(Œº={-2, 0, 2}, œÉ=1)")
plt.title("Standard Normal Distribution")
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y1 = standard_normal(x)
y2 = normal(x, 0, 0.5)
y3 = normal(x, 0, 2)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel("x")
plt.ylabel("$ùí©(Œº={-2, 0, 2}, œÉ=1)$")
plt.title("Standard Normal Distribution")
plt.show()
```

```python
k = 10000
N = 10

mean = []

for _ in range(k):
    mean.append(np.mean([np.random.rand() for _ in range(N)]))

plt.hist(mean, bins="auto", density=True)
plt.ylim(0, 5)
plt.show()
```

```python
k = 10000
N = 5

sums = []

for _ in range(k):
    sums.append(np.sum([np.random.rand() for _ in range(N)]))

x_norm = np.linspace(-5, 5, 1000)
mu = N / 2
sigma = np.sqrt(N / 12)
y_norm = normal(x_norm, mu, sigma)

plt.plot(x_norm, y_norm, label="Normal Distribution")
plt.hist(sums, bins="auto", density=True, alpha=0.5, label="Sum of Uniforms")
plt.xlim(-1, 6)
plt.ylim(0, 2)
plt.show()
```
